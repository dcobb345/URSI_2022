# Tried to do this efficiently as a loop (W.I.P)

```python
sources=['RULUP','IMLUP','GWLUP','AS205A','DOAR25','WAOPH6','FZTAU','TWCHA','VZCHA','GQLUP','GKTAU','RYLUP']

linefluxes=[]

for s in sources:
    irs=read_irs(s,dd='/Users/dougcobb/URSI_2022/REDUCED/',noerror=True)
    wave_data = pd.DataFrame([15.17], columns=['wave'])
    lineflux_data=calc_fluxes(irs.wave,irs.flux,wave_data,v_dop=200,fwhm_v=500.,sep_v=2500.,cont=0,vet_fits=True, plot=True,ymax=0.05, ymin=-0.01,amp=0.02)
    linefluxes.append(lineflux_data['lineflux'])

print(linefluxes)
```
