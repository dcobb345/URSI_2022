```python
data = pd.read_csv (r'C:\Users\pheretz\Desktop\URSI Internship\Sources Info Sheet - Catalog.csv')
df = pd.DataFrame(data, columns= ['Star','Log_Mstar','log_Teff'])
print (df)
```

```python
#Log M_star vs. Log T_eff

star_list = df['Star'].tolist()

log_Mstar = df['Log_Mstar'].tolist()

log_Teff = df['log_Teff'].tolist()
```

```python
fig=plt.figure(1,figsize=(18,10))
ax1=fig.add_subplot(111)
ax1.plot(log_Mstar, log_Teff, 'ro')
for i, (x, y) in enumerate(zip(log_Mstar, log_Teff)):
    ax1.text(x,y,star_list[i],fontsize=8)
ax1.set_xlabel('Log Mstar (M*)',fontsize=18)
ax1.set_ylabel('Log Teff (K)',fontsize=18)

#ax1.set_xlim(1,4)   #Set limits for x axis to make the plot look better
#ax1.set_ylim(90,400)
```
